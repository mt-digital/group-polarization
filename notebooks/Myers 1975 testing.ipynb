```python
def run_myers1975_trials(alpha, n_iter=100, n_trials=100):
    
    def _run_trial():
        experiment = Myers1975()
#         print(n_iter)
#         print(alpha)
        return experiment.run(n_iter, alpha=alpha).shift
    
    trials = np.zeros((n_trials,))
#     return np.mean([_run_trial() for _ in range(n_trials)])
    for trial_idx in range(n_trials):
        trials[trial_idx] = _run_trial()
    return trials
    
n_trials = 100
n_iter = 50
alphas = np.arange(0.4, 1.8, 0.2)
shifts_over_alpha = np.zeros((len(alphas), n_trials))

for idx, alpha in enumerate(alphas):
    print('Running alpha =', alpha)
    shifts_over_alpha[idx] = run_myers1975_trials(alpha, n_iter, n_trials)

# import matplotlib.pyplot as plt
# plt.plot(alphas, mean_shifts)

means_over_alpha = shifts_over_alpha.mean(axis=1)
stddev_over_alpha = shifts_over_alpha.std(axis=1)
print(len(means_over_alpha))
print(means_over_alpha)
print(stddev_over_alpha)

plt.errorbar(x=alphas, y=means_over_alpha, yerr=stddev_over_alpha)
plt.grid('y')
```
